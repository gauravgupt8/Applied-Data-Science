```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\gaurav\Desktop\Applied Data Science Github\Applied-Data-Science\kidney_disease.csv"
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
# 1. Histogram for 'age' column
plt.figure(figsize=(8, 6))
sns.histplot(data['age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Scatter plot for 'bp' (blood pressure) vs 'bu' (blood urea)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='bp', y='bu', color='green')
plt.title('Blood Pressure vs Blood Urea')
plt.xlabel('Blood Pressure (bp)')
plt.ylabel('Blood Urea (bu)')
plt.show()
```

```python
# 3. Correlation heatmap for numerical columns
# Selecting only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=['float64', 'int64']).drop(columns=['id'])

# Plotting the correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
